# 데이터 정제
## - 불필요한 삭제
## - 잘못된 행동 수정

```python
import pandas as pd

df = pd.read_csv('./namsan.csv', encoding = 'euc kr', low_memory = False)
```

```python
print(df.columns)
mask = df.columns != 'Unnamed: 13'
print(mask)
ns_df = df.loc[:, mask]    # 열을 실제로 삭제하는 것보다는 '숨기기'에 가까움.
ns_df.head(3)
```

```python
# 실제 Column 삭제하는 법
df.drop(['부가기호', 'Unnamed: 13'], axis = 1)

# 원본에서 삭제하기
df.drop(['부가기호', 'Unnamed: 13', '주제분류번호'], axis = 1, inplace = True)
```

```python
# Col 중에 하나라도 NaN(비어있음)이면 해당 col 삭제
book_df = df.dropna(axis=1)
book_df.head(3)

# col의 모든 데이터가 NaNd이면 해당 col 삭제
book_df = df.dropna(axis=1, how='all')
book_df.head(3)
```
